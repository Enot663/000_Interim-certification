#### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

- строим матрицу корреляции

````python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
````

##### Отображаем полученную матицу в виде графика

````python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
````

##### В столбце price определяем значения с наибольшей корреляцией, и с помощью графиков и диаграмм проверяем что так влияет на стоимость недвижимости.

### Конструкция и дизайн дома.

````python
v = df['grade'].value_counts()
v
````

````python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg')
plt.xlabel('Цена')
plt.ylabel('Значения конструкции и дизайна');
````

##### Из графика видим что чем выше оценка конструкции и дизайна дома, тем выше и его стоимость. Проверим данные с помощью диаграммы boxplot

````python
plt.figure(figsize=(16, 8))
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=3.5)
plt.xlabel('Цена')
plt.ylabel('Значения качества')
plt.title('Зависимость цены от качества конструкции и дизайна');
````

##### Данную диаграмму используют для быстрой оценки данных. В данном случае она показывает что дома со значением конструкции и дизайна 13 имеют наибольшую стоимость.  

### Этажность домов

````python
v = df['floors'].value_counts()
v
````

````python
sns.jointplot(x=df['price'], y=df['floors'], kind='reg')
plt.xlabel('Цена')
plt.ylabel('Количество этажей');
````

##### Исходя из данного графика мы видим что этажность дома тоже влияет на его стоимость.

````python
plt.figure(figsize=(16, 8))
sns.boxplot(x=df['price'], y=df['floors'].astype('str'),showfliers=False)
plt.xlabel('Цена')
plt.ylabel('Этажность')
plt.title('Зависимость цены от количества этажей');
````

##### Проверив версию на boxplot убедильсь что наиболее высокую цену имеют дома в 2.5 этажа.
